```python
from qiskit import *
```

```python
# For Jupyter Notebooks:
%config InlineBackend.figure_format = 'svg' # Makes the images look nice
```

```python
qc = QuantumCircuit()
```

```python
qr = QuantumRegister(2,'qreg')
```

```python
qc.add_register( qr )

qc.qregs
```

```python
qc.draw(output='mpl')
```

```python
qc.h(qr[0])
```

```python
qc.cx(qr[0], qr[1]);
```

```python
qc.draw(output='mpl')
```
